```python
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator

logdir = "outputs/dpo_m1_combined_clean_20250608_1531/runs"
ea = EventAccumulator(logdir)
ea.Reload()

print("Available scalar tags:")
for tag in ea.Tags()['scalars']:
    print("-", tag)
```

```python
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator
import matplotlib.pyplot as plt

# Path to your tensorboard log folder (adjust if needed)
logdir = "outputs/dpo_m1_combined_clean_20250608_1531"

ea = EventAccumulator(logdir)
ea.Reload()

# Get validation loss or change to any scalar name like 'train/loss'
scalars = ea.Scalars("eval/validation_loss")

steps = [s.step for s in scalars]
values = [s.value for s in scalars]

plt.plot(steps, values)
plt.xlabel("Step")
plt.ylabel("Validation Loss")
plt.title("DPO Training Curve")
plt.grid(True)
plt.show()
```
