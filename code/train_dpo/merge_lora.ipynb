```python
from transformers import AutoTokenizer, AutoModelForCausalLM
from peft import PeftModel
import torch

# === CONFIGURATION ===
base_model_id = "Qwen/Qwen3-0.6B-Base"
lora_checkpoint = "ciacco/dpo_checkpoints_unsloth_dpo_large_combined_20250609_1741"
repo_id = "ciacco/merged_qwen3_dpo_model"

# === Load base model in bfloat16 ===
print("Loading base model...")
base_model = AutoModelForCausalLM.from_pretrained(
    base_model_id,
    torch_dtype=torch.bfloat16,
    device_map="auto",
    trust_remote_code=True
)

# === Load tokenizer ===
tokenizer = AutoTokenizer.from_pretrained(base_model_id, trust_remote_code=True)

# === Load LoRA adapter ===
print("Loading LoRA adapter...")
model = PeftModel.from_pretrained(
    base_model,
    lora_checkpoint,
    torch_dtype=torch.bfloat16,
    device_map="auto",
    trust_remote_code=True
)

# === Merge adapter into base ===
print("Merging LoRA into base...")
merged = model.merge_and_unload()

# === Save and push merged model ===
print(f"Saving and pushing to Hugging Face Hub ({repo_id})...")
merged.save_pretrained(repo_id, safe_serialization=True)
tokenizer.save_pretrained(repo_id)

merged.push_to_hub(repo_id, safe_serialization=True, private=True)
tokenizer.push_to_hub(repo_id, private=True)

print(f"✅ Merge complete and pushed to: https://huggingface.co/{repo_id}")

```

```python
# === Environment Setup ===
import os
os.environ["CC"] = os.path.expanduser("~/gcc_wrapper")
os.environ.pop("GCC_EXEC_PREFIX", None)

from transformers import AutoTokenizer, AutoModelForCausalLM
from peft import PeftModel
import torch

# === CONFIGURATION ===
base_model_id = "matteodagos/MNLP_M3_mcqa_model"
lora_checkpoint = "ciacco/dpo_checkpoints_unsloth_dpo_large_combined_20250609_1741"
repo_id = "ciacco/MNLP_M3_mcqa_dpo_model_alpha_05"
alpha = 0.5  # Interpolation weight: 1.0 = pure DPO, 0.0 = pure base

# === Load base model ===
print("🔁 Loading base model...")
base_model = AutoModelForCausalLM.from_pretrained(
    base_model_id,
    torch_dtype="auto",  # Or torch.bfloat16
    device_map="auto",
    trust_remote_code=True,
)

# === Load tokenizer ===
tokenizer = AutoTokenizer.from_pretrained(base_model_id, trust_remote_code=True)

# === Load LoRA into base model ===
print("🔗 Loading DPO LoRA adapter...")
model = PeftModel.from_pretrained(
    base_model,
    lora_checkpoint,
    torch_dtype="auto",
    device_map="auto",
    trust_remote_code=True
)

# === Interpolate adapter using `add_weighted_adapter` ===
print(f"🧪 Interpolating adapter with alpha={alpha}...")
model.add_weighted_adapter(
    adapters=["default"],   # your current LoRA adapter is called "default"
    weights=[alpha],        # how much of it to blend in
    adapter_name="interpolated_dpo",
    combination_type="linear"
)

model.set_adapter("interpolated_dpo")

# === Merge adapter into base model ===
print("🧬 Merging interpolated adapter into base...")
merged = model.merge_and_unload()

# === Save and push to Hugging Face Hub ===
print(f"💾 Saving and pushing to Hugging Face Hub ({repo_id})...")
merged.save_pretrained(repo_id, safe_serialization=True)
tokenizer.save_pretrained(repo_id)

merged.push_to_hub(repo_id, safe_serialization=True)
tokenizer.push_to_hub(repo_id)

print(f"✅ Merge complete and pushed to: https://huggingface.co/{repo_id}")
```
